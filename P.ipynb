```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('tube1.csv')
filtered_df = df[(df['f3'] == 0) & (df['f4'] == 0) & (df['f5'] == 0.875)]
power = filtered_df['f7'] * filtered_df['f8']

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(power, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Power")
plt.xlabel("Power")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(power, bw_adjust=0.5)
plt.axvline(0.1, color='orange', linestyle='--')
plt.axvline(0.3, color='orange', linestyle='--')
plt.title("KDE of Power")
plt.show()
```

```python
low = power[power <= 0.1]
mid = power[(power > 0.1) & (power <= 0.3)]
high = power[power > 0.3]

print(f"Total samples: {len(power)}")
print(f"Power  0.1: {len(low)} ({len(low)/len(power)*100:.2f}%)")
print(f"0.1 < Power  0.3: {len(mid)} ({len(mid)/len(power)*100:.2f}%)")
print(f"Power > 0.3: {len(high)} ({len(high)/len(power)*100:.2f}%)")
```

```python
high_power_f2 = filtered_df.loc[power > 0.3, 'f2']

print(high_power_f2)
```

```python
power = filtered_df['f7'] * filtered_df['f8']
high_power_f2 = filtered_df.loc[power > 0.2, 'f2']
print(high_power_f2)
```

```python
power = filtered_df['f7'] * filtered_df['f8']
high_power_f2 = filtered_df.loc[power > 0.1, 'f2']

print(high_power_f2)
```

```python

power_rounded = power.round(4)  # rounding to 4 decimals
top_3_power = power_rounded.value_counts().head(3)

print(top_3_power)
```
